# Cluster analysis
In this notebook we explore heatmaps and dendrograms

## Import the relevant libraries

```python
import pandas as pd
import seaborn as sns

sns.set_theme()
```

## Load the data

```python
data = pd.read_csv('Country clusters standardized.csv', index_col='Country')
data.head()
```

```python
x = data.copy()
x = x.drop(['Language'], axis= 1)
x.head()
```

## Plot the data

```python
sns.clustermap(x, cmap='mako')
```

## Interpretation

Latitude:
- Most countries have similar latitudes since they are in the northern hemisphere, the only difference is Australia which is in the southern one, hence such different colour from the other ones  

Longitude:
- Here we can observe 3 different colour groups and it is due to the fact that North America is on the left side of our map, Europe is in the centre and Oceania is on the right
