# The histogram

**Background:** You are given a dataset.<br>
**Task:** Create a histogram with 10 intervals, based on your dataset.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [13, 68, 165, 193, 216, 228, 361, 470, 500, 529, 544, 602, 647, 692, 696, 699, 809, 892, 899, 936]

# Has to be the size of desired intervals+1 because of the way linspace works
bins = np.linspace(min(data), max(data), 10 + 1)

plt.hist(data, bins=bins)
plt.xticks(bins)
plt.show()
```
