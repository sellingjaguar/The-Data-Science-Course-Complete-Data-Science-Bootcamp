# Categorical variables. Visualization techniques
## Ice cream shop
Note: You may solve these problems both on paper or in a software of your choice. The medium is not crucial.<br><br>
**Background:** There is an ice cream shop, that is operating in New York, LA and San Francisco.<br>
**Data:** You have sold 12,327 ice creams in New York; 17,129 in LA and 19,923 in San Francisco.<br>
**Task:** Order the data in a frequency distribution table.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

# Get the values
locations = ['New York', 'LA', 'San Francisco']
values = [12327, 17129, 19923]

# Create pandas DF
df = pd.DataFrame({'location': locations, 'units_sold': values})

# Base data frame
df
```

```python
# Pie Chart with percentage
plt.pie(df['units_sold'], labels=df['location'], autopct='%0.2f%%')
plt.show()
```

```python
# Pareto Chart

# Order and add cumulative percentage
pareto_data = df.sort_values(by='units_sold', ascending=False)
pareto_data['cum_perc'] = pareto_data['units_sold'].cumsum()/pareto_data['units_sold'].sum()*100

# Make subplot since we will have 2 plots together
fig, ax = plt.subplots()
plt.title('Units sold')

# Plot units_sold values
ax.bar(pareto_data['location'], pareto_data['units_sold'])

# Plot cumulative percentage
ax2 = ax.twinx()
ax2.plot(pareto_data['location'], pareto_data['cum_perc'], color='red', marker='D')
ax2.yaxis.set_major_formatter(PercentFormatter())
ax2.tick_params(axis='y', colors='red')

plt.show()
```
