# Context
We have data about customers and wether or not they are premium members, the data consists of age and premium status.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('Customers_Membership.xlsx')
```

```python
age = data['Customer Age ']
status = data['Membership Status']

plt.scatter(age, status)
plt.xlabel('Age')
plt.ylabel('Status')
plt.show()
```

In this plot we can verify that the older members tend to be much more likely to have a premium membership than the young ones and teh division is around 34-35 years old, this could indicate that this is the age where they are more likely to have reached the financial stability necessary to be able to dispose of their income in a service like this.

```python
young_premium_count = np.sum((data['Customer Age '] <= 34) & (data['Membership Status'] == 1))
percent_young_premium = young_premium_count / np.sum(data['Customer Age '] <= 34)

old_premium_count = np.sum((data['Customer Age '] > 34) & (data['Membership Status'] == 1))
percent_old_premium = old_premium_count / np.sum(data['Customer Age '] > 34)

print(f'The percentage of young members that are premium is {percent_young_premium * 100:.2f}% and they are {young_premium_count} in total')
print(f'The percentage of old members that are premium is {percent_old_premium * 100:.2f}% and they are {old_premium_count} in total')
```

As proven by the statistics above not only the percentage of older members that have premium memberships is higher, but also their total number is higher, this undoubtedly tells us that we should target their demographic should we run any marketing campaigns.
