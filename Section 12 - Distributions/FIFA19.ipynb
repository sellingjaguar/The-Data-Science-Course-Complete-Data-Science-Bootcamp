```python
# Do all the imports and get the data
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('FIFA19.csv')

def range(arr):
    return int(max(arr) - min(arr))
```

```python
# Make histogram overall

overall = data['Overall']

plt.hist(overall, bins=range(overall))
plt.xlabel('Overall score')
plt.ylabel('Number of players')
plt.show()
```

We can verify that the Overall score follows a normal distribution hence the players in the game appear to be balanced.
Now lets try with a random sample:

```python
overall_sample = overall.sample(60)

plt.hist(overall_sample, bins=range(overall_sample))
plt.xlabel('Sample overall score')
plt.ylabel('Amount')
plt.show()
```

We can see that now with a small sample instead of following a normal distribution it follows a students' T distribution which is a common effect when selecting small samples from data that follows a normal distribution.     
Now let's try with an individual stat:

```python
penalties = data['Penalties']

plt.hist(penalties, bins=range(penalties))
plt.xlabel('Penalty score')
plt.ylabel('Amount')
plt.show()
```

Here we have 2 interesting characteristics that could be pointed out:
1. It seems that there are 2 groups of players which I would categorize as non-viable (left "mountain") or viable (right "mountain") for penalties, if I had to assume, the non-viable players are composed of the goal keepers and defense while the viable ones are the midfielders and the attackers.
2. There seem to be a lot of average players regarding penalty shooting ability and only a small number is good, yet, the difference between good and bad is very noticeable meaning that there are a lot more bad than good.

Let's try to check characteristic 1:

```python
gk_handling = data['GKHandling']

plt.hist(gk_handling, bins=range(gk_handling))
plt.xlabel('GK handling score')
plt.ylabel('Amount')
plt.show()
```

Here we can observe that there is an even clearer separation between players which I would categorize as goal keeper and non goal keeper, as we can observe the left cluster is made up of most of the population, this is due to the fact that on the left side we have every player that isn't a goal keeper.   
Now let's check the ages of the players and see if we can find anything interesting:


```python
age = data['Age']

plt.hist(age, bins=range(age))
plt.xlabel('Age')
plt.ylabel('Amount')
plt.show()
```

Here we can see that the age follows a Poisson distribution, meaning that the young players outweigh the old ones, we can also see that the minimum age is 16, this is probably due to the fact that FIFA19 only has first team players and that the minimum age to play professionally in those teams is 16, we can also see that most players are between 21 and 27 years old and after that the amount of players that keep playing declines continuously, this is probably due to retirement and injuries.
