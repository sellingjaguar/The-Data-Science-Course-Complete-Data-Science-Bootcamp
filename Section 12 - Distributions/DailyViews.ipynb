# Context
A company posts a video every week on monday and records the number of daily views, this is the data for one of those videos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('Daily+Views.xlsx')
```

```python
days = data['Days after release']
views = data['Daily Views']

plt.plot(days, views, 'o')
plt.xlabel('Days after publishing')
plt.ylabel('Views (x1000)')

log_views = np.log(views)
coeffs = np.polyfit(days, log_views, 1)
poly = np.poly1d(coeffs)
trendline = np.exp(poly(days))

plt.plot(days, trendline, ':')

plt.show()
```

We can observe that this data follows an exponential distribution, we can also observe how the views start of strong and eventually reach a turning point around day 8, this is probably due to the fact that the video is not recent with the addition of the release of a new video that shifts attention, after this point it plateaus with a low view count.

```python

pdf = [v/sum(views) for v in views]
cdf = np.cumsum(pdf)

plt.plot(days, pdf, ':', color="red", label="PDF") 
plt.plot(days, cdf, '-', color="blue", label="CDF") 

plt.legend()
```

By plotting the PDF and the CDF together we can verify that they also follow what is expected of a exponential distribution.   
This analysis served to show as the the impact of a campaign like this has a very short term impact, in this case this could be done close to the release of the marketed product to generate interest until the launch date.
